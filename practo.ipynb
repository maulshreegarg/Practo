```python
import pandas as pd
data=pd.read_csv("/content/ent.csv")
df = pd.DataFrame(data)
df.head()
```

```python
df.head()
print(pd.isna(df))
df.info()
```

```python
df["price"]=df["price"].str.replace('fee','')
```

```python
df['price']=pd.to_numeric(df["price"])
```

```python
df.head()
```
